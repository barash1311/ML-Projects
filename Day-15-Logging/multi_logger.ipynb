```python
import logging

##creating a module1 logger
logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

##create a logger for module2
logger2=logging.getLogger("module 2 ")
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s-%(name)s-%(levelname)s-%(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)
```

```python
#log messages
logger1.debug("this is a debug messsage for moduul1")
logger2.warning("this is warning of the module 2")
logger2.error("this is the error for the module ")
```
