```python
import numpy as np
```

```python
list_of_marks=[45,23,23,4,5,67,89,76,54,32,55]
minimum,q1,median,q3,maxinum=np.quantile(list_of_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,q1,median,q3,maxinum
```

```python
IQR=q3-q1
print(IQR)
```

```python
lower_fence=q1-1.5*IQR
high_fence=q3+1.5*IQR
```

```python
lower_fence,high_fence
```

```python
import seaborn as sns
sns.boxplot(list_of_marks)
```
