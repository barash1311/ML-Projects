```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
X=pd.DataFrame(iris["data"],columns=["sepal length in cm\n","sepal width in cm\n"," petal length in cm\n"," petal width in cm\n"])##independent feature
y=iris["target"]##dependent feautres
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#decision tree do not need any strandradixation at all
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
#visualize the decision tree classfier
from sklearn import tree
plt.Figure(figsize=(12,8))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
##hyper parameter tuning
params={
    "criterion":["gini","entropy", "log_loss"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5],
    "max_features":["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
treemodel=DecisionTreeClassifier()
grid=GridSearchCV(treemodel,params,n_jobs=-1,cv=5,scoring="accuracy")
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
