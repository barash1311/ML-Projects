```python
#clean the data of date of journey
df["Date"]=df["Date_of_Journey"].str.split("/").str[0]
df["Month"]=df["Date_of_Journey"].str.split("/").str[1]
df["Year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.info()
#here we can see the three object that we have conversted are still the object type so we have to convert it thet he object type
```

```python
df["Date"]=df["Date"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
#now we have converted tehe data type into the int from the object
```

```python
df.info()
```

```python
#now as we have converted the date_of _journy to the differnt column we can drop the current column
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
#now lets go with the other paramteresti do the things like arrival time
#we have to split the hours and the time

df["arrival_hour"]=df["Arrival_Time"].str.split(":").str[0]
df["arrival_minutes"]=df["Arrival_Time"].str.split(":").str[1]
df["departing_hour"]=df["Dep_Time"].str.split(":").str[0]
df["departing_minutes"]=df["Dep_Time"].str.split(":").str[1]
```

```python
import pandas as pd
data=df["Duration"].str.split(" ")
def convert_to_minutes(duration):
    parts=duration.split()
    hours=0
    minute=0
    for part in parts:
        if "h" in part:
            hours=int(part.replace("h",""))
        elif "m" in part:
            minute=int(part.replace("m",''))
    
    total_minutes=hours*60+minute
    return total_minutes
df["Duration"]=df["Duration"].apply(convert_to_minutes)
```
